```python
import pandas as pd
data = pd.read_csv("/Users/sripathibhairu/Downloads/frailty_data.csv")
print(data)
```

```python
data.info()
```

```python
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns
```

```python
sns.lineplot(x='Age', y='Grip strength', data=data)
plot.title('Age Vs Grip Strength')
plot.xlabel('Age')
plot.ylabel('Grip Strength')
plot.savefig("/Users/sripathibhairu/Downloads/agevsgrip.png")
plot.show()
```

```python
sns.scatterplot(x='Height', y='Grip strength', data=data)
plot.title('Height Vs Grip Strength')
plot.xlabel('Height')
plot.ylabel('Grip Strength')

# Save the plot as a PNG file
plot.savefig(r"/Users/sripathibhairu/Downloads/heightvsgrip.png")

# Display the plot
plot.show()
```

```python
sns.distplot(data['Age'], kde = False, color ='green', bins = 50)
fig = plot.gcf()
fig.savefig(r"/Users/sripathibhairu/Downloads/Age_Distribution.png")
```
