# Student Performance Analysis - Data Visualization

This notebook includes five different visualizations analyzing student performance.

## 1. Load Dataset
We first load the dataset and inspect its structure.

```python

import pandas as pd

# Load dataset
df_students = pd.read_csv("StudentsPerformance.csv")

# Display first five rows
df_students.head()
```

## 2. Math Score Distribution
This histogram helps us understand the spread and distribution of math scores.

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Math Scores
plt.figure(figsize=(8, 5))
sns.histplot(df_students['math score'], bins=20, kde=True, color="blue")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
plt.savefig("/Users/sripathibhairu/Downloads/Distribution of Math .png")
```

## 3. Gender vs. Average Scores
This bar chart compares average scores in different subjects between male and female students.

```python

# Bar chart for average scores by gender
df_students.groupby("gender")[["math score", "reading score", "writing score"]].mean().plot(kind="bar", colormap="coolwarm", figsize=(8,5))
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")

plt.show()
plt.savefig("/Users/sripathibhairu/Downloads/Average_Score.png")
```

## 4. Parental Education Level vs. Math Scores
This box plot shows how parental education influences student math scores.

```python

# Box plot for math scores based on parental education
plt.figure(figsize=(10, 6))
sns.boxplot(x="parental level of education", y="math score", data=df_students, palette="Set2")
plt.xticks(rotation=45)
plt.title("Math Scores by Parental Education Levell")
plt.show()
plt.savefig("/Users/sripathibhairu/Downloads/Parent_vs_math.png")
```

## 5. Test Preparation Course vs. Math Scores
This violin plot visualizes the impact of test preparation courses on student performance.

```python

# Violin plot for math scores based on test preparation course completion
plt.figure(figsize=(8, 5))
sns.violinplot(x="test preparation course", y="math score", data=df_students, palette="pastel")
plt.title("Math Scores vs. Test Preparation Course")
plt.show()
plt.savefig("/Users/sripathibhairu/Downloads/Math_vs_Test.png")
```

## 6. Correlation Heatmap of Scores
This heatmap shows relationships between scores in different subjects.

```python

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_students[["math score", "reading score", "writing score"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Scores")
plt.show()
plt.savefig("/Users/sripathibhairu/Downloads/Correlation_Heatmap.png")
```
